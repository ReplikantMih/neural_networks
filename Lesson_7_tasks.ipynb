##### Review:

Данная нейросеть предполагается для работы на мобильных устройствах (что накладывает соответствующие ограничения по тяжеловесности и производительности).

Модель MnasFPN совместно с MobileNetV2 выигрывает у комбинации MobileNetV3+SSDLite 1.8 единицы по показателю mAP при одинаковом времени выполнения. Или данная комбинация на 1 mAP лучше и на 10% быстрее, чем NAS-FPNLite.

Архитектура разработана с применением Neural architecture
search (NAS) – технология автоматизация разработки архитектуры нейросети (с применением принципов обучения с подкреплением). 

В данной нейросети была переработана только блок детектирования, бэкбон был взят готовый.

При разработке использовался фреймворк для NAS, поддерживающий в качестве параметра величину задержки (время на анализ изображения), что позволило на выходе получить архитектуру детектирующей части, оптимизированную с точки зрения этого параметра в том числе.

Типичным для цели оптимизации задержки сети является использование легковесного бэкбона, данная же команда пошла по пути оптимизации скорости работы именно детектирующей части.

В процессе исследования были выбраны несколько оптимизированных NAS по параметру задержки моделей. По анализу результатов модификаций моделей была выбрана оптимальная архитектура.

Итоговая архитектура содержит 2 инновации, позволившие добиться обозначенных ранее результатов в оптимизации с точки зрения соотношения производительность-качество:
1.	Добавление блоков инвертированных остатков в часть нейросети, отвечающую за детектирование.
2.	Были перестроены операции по изменению изменении размерности слоев для организации эффективного объединения информации между масштабами.
Прирост скорости относительно сетей, с которыми идет сравнение, за счет только детектирующей части сети составил 25% (при неизменном качестве). Прирост настолько значительный, что сделал бэкбон бутылочным горлышком по производительности в итоговой модели (до этого было наоборот).
